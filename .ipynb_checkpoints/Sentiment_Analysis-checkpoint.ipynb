```python
import re
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
# nltk.download('stopwords')
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
from tqdm import tqdm
from nltk.corpus import stopwords
stopwordEn = stopwords.words('english')
import pickle
import os

import multiprocessing as mp
import numpy as np
import time
print('CPU numbers:',mp.cpu_count())
def _apply_df(args):
    df, func, kwargs = args
    return df.apply(func, **kwargs)

def apply_by_multiprocessing(df, func, **kwargs):
#     print(kwargs)
    workers = kwargs.pop('workers')
    pool = mp.Pool(processes=workers)
    result = pool.map(_apply_df, [(d, func, kwargs) for d in np.array_split(df, workers)])
    pool.close()
    return pd.concat(list(result))
#apply_by_multiprocessing(fullset['Text'], process_text, workers=cores)

def processText(text,lemma=False, gram=1):
    text = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b|@\w+|#', '', text, flags=re.MULTILINE) #delete URL, #hashtag# , and @xxx
    tokens = word_tokenize(text)
    whitelist = ["n't", "not", "no"]
    tokens = [lemmaWord(i.lower()) if lemma else i.lower() for i in tokens if (i.lower() not in stopwordEn or i.lower() in whitelist) and i.isalpha()]
    if gram<=1:
        return tokens
    else:
        return [' '.join(i) for i in nltk.ngrams(tokens, gram)]
```

# Tokenize and Encode Text

```python
t = Tokenizer(num_words=None) #设置为None则使用全部词语，若设置数字N，则在后续Text2Seq时只考虑TopN的词汇
t.fit_on_texts(X_train) # 根据给出的text训练tokenizer, 本质上是构建一个记录词频的字典，仅仅在X_train上训练，后续测试若有新词则跳过
max_length = X_train.map(len).max()  # 21

print(list(t.word_index.keys())[:10]) #自动根据词频高到低排序，index从小到大，这里显示所有的词语，与Tokenizer的num_words无关

vocab_size = len(t.word_index) + 1
print(f"Max length:{max_length}, Vocab size:{vocab_size}")
```

## Text to sequence

```python
# 不用Embedding的转变文字为向量的方法，通过使用固定长度的word_dict词频字典，把文字转换为对应的字典index
# 然后再将这个index转换为one-hot编码，得到矩阵，这里会丢失同一词语在一个doc里多次出现的信息，次顺序也会丢失
docs_seq = t.texts_to_sequences(X_train)  #根据word_index字典把对应的word转换为index
docs_seq_test = t.texts_to_sequences(X_test)

def one_hot_seq(seqs, nb_features = vocab_size):
    ohs = np.zeros((len(seqs), nb_features))
    for i, s in enumerate(seqs):
        ohs[i, s] = 1.
    return ohs
# Size = rows * vocab size
docs_oh = one_hot_seq(docs_seq)
docs_oh_test = one_hot_seq(docs_seq_test)
# from sklearn.preprocessing import LabelEncoder
# y_train_le = le.fit_transform(df.airline_sentiment)
# y_train_oh = to_categorical(y_train_le)

print('"{}" is converted into {}'.format(docs_seq[0], docs_oh[0]))
print('For this example we have {} features with a value of 1. out of {} unique words'.format(docs_oh[-1].sum(),len(docs_oh[-1])))
```

## Pre-Trained GloVe Embedding
https://nlp.stanford.edu/projects/glove/

```python
# load the whole embedding into memory
EMB_DIM = 25
embeddings_dict = {}
with open(F"../Pretrained_WE/glove.twitter.27B.{EMB_DIM}d.txt", 'r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        if len(vector)==EMB_DIM:
            embeddings_dict[word] = vector
print('Loaded %s word vectors.' % len(embeddings_dict))
```

```python
emb_matrix = np.zeros((vocab_size, EMB_DIM))

for w, i in t.word_index.items():
    # The word_index contains a token for all words of the training data so we need to limit that
    if i < vocab_size:
        vect = embeddings_dict.get(w)
        # Check if the word from the training data occurs in the GloVe word embeddings
        # Otherwise the vector is kept with only zeros
        if vect is not None:
            emb_matrix[i] = vect
    else:
        break
emb_matrix.shape
```

## Word Embedding FC Model

```python
def we_model(EMB_DIM=25, OUTPUT_ACT='softmax',LOSS='categorical_crossentropy'):
    tf_random.set_seed(0)
    model = Sequential()
    model.add(Embedding(vocab_size, EMB_DIM, input_length=max_length))
    model.add(Flatten())
    model.add(Dropout(0.5))
    model.add(Dense(16, activation='relu',kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform'))
#     model.add((BatchNormalization()))
    model.add(Dropout(0.5))
    model.add(Dense(3, activation=OUTPUT_ACT,kernel_regularizer=regularizers.l2(0.001)))
    model.compile(optimizer='adam', loss=LOSS, metrics=['acc'])
    return model
model_we = we_model()
model_we.summary()
```

```python
def geLSTM_model(EMB_DIM=25, OUTPUT_ACT='softmax',LOSS='categorical_crossentropy',addRelu=False,return_sequences=False):
    tf_random.set_seed(0)
    model = Sequential()
    model.add(Embedding(vocab_size, EMB_DIM, input_length=max_length,weights=[emb_matrix],trainable=True))
    model.add(Dropout(0.5))
    model.add(LSTM(16, return_sequences=return_sequences, kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform')) 
    if addRelu: model.add(Activation('relu'))
    model.add(Flatten())
    model.add(Dropout(0.5))
    model.add(Dense(3, activation=OUTPUT_ACT, kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform'))
    model.compile(optimizer='adam', loss=LOSS, metrics=['acc'])
    return model
```

```python
def weCNN_model(EMB_DIM=25, OUTPUT_ACT='softmax',LOSS='categorical_crossentropy',CNNsize=2, pool=None,dropFirst=True):
    tf_random.set_seed(0)
    model = Sequential()
    model.add(Embedding(vocab_size, EMB_DIM, input_length=max_length))
    model.add(Dropout(0.5))
    model.add(Conv1D(16, CNNsize, padding='valid', activation='relu', strides=1,kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform'))
    if dropFirst: model.add(Dropout(0.5))
    if pool=='GlobalMax':
        model.add(GlobalMaxPooling1D())
    elif pool=='Max':
        model.add(MaxPooling1D(pool_size=2, strides=None, padding="valid", data_format="channels_last")) # 必须接flatten
    if dropFirst!=True: model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(3, activation=OUTPUT_ACT))
    model.compile(optimizer='adam', loss=LOSS, metrics=['acc'])
    return model
model_weCNN = weCNN_model()
model_weCNN.summary()
```

## Glove Embedding CNN model

```python
def ge_cnn_model(EMB_DIM=25,LOSS='categorical_crossentropy',CNNsize=2, pool=None,dropFirst=True,trainable=True):
    tf_random.set_seed(0)
    model = Sequential()
    model.add(Embedding(vocab_size, EMB_DIM, input_length=max_length,weights=[emb_matrix],trainable=trainable))
    model.add(Dropout(0.5))
    model.add(Conv1D(16, CNNsize, padding='valid', activation='relu', strides=1,kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform'))
    if dropFirst: model.add(Dropout(0.5))
    if pool=='GlobalMax':
        model.add(GlobalMaxPooling1D())
    elif pool=='Max':
        model.add(MaxPooling1D(pool_size=2, strides=None, padding="valid", data_format="channels_last")) # 必须接flatten
    if dropFirst!=True: model.add(Dropout(0.5))
    model.add(Flatten())
#     model.add(Dense(16, activation='relu',kernel_regularizer=regularizers.l2(0.001),kernel_initializer='he_uniform'))
#     model.add(Dropout(0.5))
    model.add(Dense(3, activation='softmax'))
    model.compile(optimizer='adam', loss=LOSS, metrics=['acc'])
    return model
```
